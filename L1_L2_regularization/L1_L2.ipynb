```python
import pandas as pd
```

```python
df = pd.read_csv('melb_data.csv')
```

```python
df.head(20)
```

```python
df.columns
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df['Car']= df['Car'].fillna(0)
```

```python
df.isna().sum()
```

```python
df['BuildingArea'] = df['BuildingArea'].fillna(df.BuildingArea.mean())
```

```python
df['YearBuilt'] = df['YearBuilt'].fillna(df.YearBuilt.mean())
```

```python
df.isna().sum()
```

```python
df=df.drop(['Postcode','Type','Method','Suburb','SellerG', 'Date'],axis=1)
```

```python
df.head()
```

```python
df.columns
```

```python
df= pd.get_dummies(df,drop_first=True,dtype=int)
df= df.astype(int)
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python

from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X=df.drop('Price',axis=1)
```

```python
X.shape
```

```python
y = df['Price']
```

```python
y.shape
```

```python
X_train ,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape
```

```python
len(X_test)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model.score(X_train,y_train)
```

```python
from sklearn import linear_model
```

```python
l1 = linear_model.Lasso(alpha=2,max_iter=800,tol=1,positive=True)
l1.fit(X_train,y_train)
```

```python
l1.score(X_train,y_train)
```

```python
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
param_grid = {
    'alpha': [0.001, 0.01, 0.1, 1, 10, 100],
    'max_iter': [1000, 2000, 5000],
    'tol': [1e-4, 1e-3, 1e-2],
    'fit_intercept': [True, False]
}
```

```python
rc = RandomizedSearchCV(l1,param_grid,cv=5,return_train_score=False,n_iter=20)
```

```python
rc
```

```python
rc.fit(X_test,y_test)
```

```python
rc.best_score_
```
